```python
data = pd.read_csv('888.csv', sep=';', decimal=',')
data.dropna(axis='columns', how='all', inplace=True)
data.head()
```

```python
columns = data.columns

new_columns = []
new_columns.append(columns[0])
new_columns.append(columns[1])

for i in range(2, len(columns)):
    a, b = columns[i].split(';')
    date = a.split(' ')
    if date[1] == 'января':
        date[1] = 1
    elif date[1] == 'февраля':
        date[1] = 2
    elif date[1] == 'марта':
        date[1] = 3
    elif date[1] == 'апреля':
        date[1] = 4
    elif date[1] == 'мая':
        date[1] = 5
    elif date[1] == 'июня':
        date[1] = 6
    elif date[1] == 'июля':
        date[1] = 7
    elif date[1] == 'августа':
        date[1] = 8
    elif date[1] == 'сентября':
        date[1] = 9
    elif date[1] == 'октября':
        date[1] = 10
    elif date[1] == 'ноября':
        date[1] = 11
    elif date[1] == 'декабря':
        date[1] = 12
    else:
        break
    date = [int(x) for x in date]
    a = datetime.date(date[2], date[1], date[0])
    new_columns.append(a.strftime("%d-%m-%Y") + b)

data =data.rename(columns=dict(zip(columns, new_columns)))
```

```python
def get_dataset_of_company(Tickers, ticker, data):
    
    data_index = data[(data['01 Tickers'] == Tickers)& (data['ticker'] == ticker )].index
    a = data.loc[data_index[0]]
    inx = a.index
    dates = []
    data_Open = []
    data_DayHigh = []
    data_DayLow = []
    data_Volume = []
    data_Close = []
    for i in range(2, len(a)):
        if 'Open' in inx[i]:
            date, _ = inx[i].split(' ')
            dates.append(date)
            data_Open.append(a[i])
        elif 'DayHigh' in inx[i]:
            data_DayHigh.append(a[i])
        elif 'DayLow' in inx[i]:
            data_DayLow.append(a[i])
        elif 'Volume' in inx[i]:
            data_Volume.append(a[i])
        elif 'Price' in inx[i]:
            data_Close.append(a[i])
    data_of_company = pd.DataFrame({'Open' : data_Open, 'DayHigh': data_DayHigh, \
              'DayLow': data_DayLow, 'Volume': data_Volume,\
              'Close': data_Close}, index=dates)
    return Tickers, ticker, data_of_company

def get_dataset_of_company_by_id(company_id, data):
    a = data.loc[company_id]
    inx = a.index
    dates = []
    data_Open = []
    data_DayHigh = []
    data_DayLow = []
    data_Volume = []
    data_Close = []
    for i in range(2, len(a)):
        if 'Open' in inx[i]:
            date, _ = inx[i].split(' ')
            dates.append(date)
            data_Open.append(a[i])
        elif 'DayHigh' in inx[i]:
            data_DayHigh.append(a[i])
        elif 'DayLow' in inx[i]:
            data_DayLow.append(a[i])
        elif 'Volume' in inx[i]:
            data_Volume.append(a[i])
        elif 'Price' in inx[i]:
            data_Close.append(a[i])
    data_of_company = pd.DataFrame({'Open' : data_Open, 'DayHigh': data_DayHigh, \
              'DayLow': data_DayLow, 'Volume': data_Volume,\
              'Close': data_Close}, index=dates)
    return a['01 Tickers'], a['ticker'], data_of_company
```

```python
import talib
```

```python
def calculate_parametres(dataset, rsi=False, stoch=False, stochrsi=False, macd=False, adx=False, willr=False):
    if rsi:
        rsi_14 = talib.RSI(dataset['Close'],  timeperiod=14)
        dataset['RSI_14'] = rsi_14
        
    if stoch:
        slowk, fastd = talib.STOCH(dataset['DayHigh'], dataset['DayLow'], dataset['Close'], fastk_period=3,\
                                   slowk_period=3, slowk_matype=0, slowd_period=3, slowd_matype=0)
        dataset['STOCH_slowk'] = slowk
        dataset['STOCH_fastd'] = fastd
        
    if stochrsi:
        fastk, fastd = talib.STOCHRSI(dataset['Close'],  timeperiod=14, fastk_period=5, fastd_period=3, fastd_matype=0)
        dataset['STOCHRSI_fastk'] = fastk
        dataset['STOCHRSI_fastd'] = fastd
        
    if macd:
        macd, macdsignal, macdhist = talib.MACD(dataset['Close'], fastperiod=12, slowperiod=26, signalperiod=9)
        dataset['MACD'] = macd
        dataset['MACD_signal'] = macdsignal
    
    if adx:
        real = talib.ADX(dataset['DayHigh'], dataset['DayLow'], dataset['Close'], timeperiod=14)
        dataset['ADX'] = real
        
    if willr:
        real = talib.WILLR(dataset['DayHigh'], dataset['DayLow'], dataset['Close'], timeperiod=14)
        dataset['WILLIAMS'] = real
```

```python
def calculate_dynamics_week(dataset, date):
    inxs = dataset.index
    date = date.split('-')
    date = [ int(x) for x in date]
    date = datetime.datetime(date[-1], date[-2], date[-3])
    
    date_1 = date
    i =7
    
    changes_in_percents = None
    
    while i > 0:
        date_2 = datetime.timedelta(days = i)
        date_2 = date_1 - date_2
        if date_2.strftime("%d-%m-%Y") in inxs:
            x = d.loc[date.strftime("%d-%m-%Y")].Close
            y = d.loc[date_2.strftime("%d-%m-%Y")].Close
            changes_in_percents = (y - x) /x *100
            break
        else:
            i -=1
            
    return changes_in_percents
```

```python
def calculate_dynamics_day(dataset, date):
    inxs = dataset.index
    date = date.split('-')
    date = [ int(x) for x in date]
    date = datetime.datetime(date[-1], date[-2], date[-3])
    
    date_1 = date
    i = 1
    
    changes_in_percents = None    
    
    while i < 7:
        date_2 = datetime.timedelta(days = 1)
        date_2 = date_1 - date_2
        
        if date_2.strftime("%d-%m-%Y") in indx:
            x = d.loc[date.strftime("%d-%m-%Y")].Close
            y = d.loc[date_2.strftime("%d-%m-%Y")].Close
            changes_in_percents = (y - x) /x *100    
            break
        else: 
            i +=1
            
    return changes_in_percents 
```

```python
def monthdelta(date, delta):
    m, y  = (date.month+ delta) %12, date.year + ((date.month) + delta-1) // 12
    if not m: m = 12
    d = min(date.day, [31, 29 if y%4==0 and not y%400==0 else 28,31,30,31,30,31,31,30,31,30,31][m-1])
    return date.replace(day=d,month=m, year=y)
```

```python
def calculate_dynamics_month(dataset, date):
    inxs = dataset.index
    date = date.split('-')
    date = [ int(x) for x in date]
    date = datetime.datetime(date[-1], date[-2], date[-3])
    
    date_1 = date
    date_2 = monthdelta(date_1, -1)
    changes_in_percents = None    

    if date_2.strftime("%d-%m-%Y") in indx:
        x = d.loc[date.strftime("%d-%m-%Y")].Close
        y = d.loc[date_2.strftime("%d-%m-%Y")].Close
        changes_in_percents = (y - x) /x *100    
    else:
        i = 1
        while i < 13: 
            date_2 = date_2  + datetime.timedelta(days = 1)
            if date_2.strftime("%d-%m-%Y") in indx:
                x = d.loc[date.strftime("%d-%m-%Y")].Close
                y = d.loc[date_2.strftime("%d-%m-%Y")].Close
                changes_in_percents = (y - x) /x *100    
                break
            else: 
                i +=1
    return changes_in_percents   
```

```python
def calculate_dynamics_year(dataset, date, years):
    inxs = dataset.index
    date = date.split('-')
    date = [ int(x) for x in date]
    date = datetime.datetime(date[-1], date[-2], date[-3])
    
    date_2 = monthdelta(date, years*12)
    changes_in_percents = None 
    if date_2.strftime("%d-%m-%Y") in indx:
        x = d.loc[date.strftime("%d-%m-%Y")].Close
        y = d.loc[date_2.strftime("%d-%m-%Y")].Close
        changes_in_percents = (y - x) /x *100    
    else:
        i = 1 
        while i < 14:
            date_2 = date_2  + datetime.timedelta(days = 1)
            if date_2.strftime("%d-%m-%Y") in indx:
                x = d.loc[date.strftime("%d-%m-%Y")].Close
                y = d.loc[date_2.strftime("%d-%m-%Y")].Close
                changes_in_percents = (y - x) /x *100    
                break
            else: 
                i +=1
    return changes_in_percents    
```
